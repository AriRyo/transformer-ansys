# デフォルト


```python
import random
import operator
import math

# 数式生成に使うオペレータと対応するPythonの演算子
operators = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
    '/': operator.truediv,
}

# ランダムな数式を生成する関数
def generate_expression(depth=0):
    if depth >= 1 and random.random() < 0.7:
        # 数字のみを返す
        return str(random.randint(1, 10))
    else:
        # 深さを増やしながら再帰的に数式を生成
        op = random.choice(list(operators.keys()))
        left = generate_expression(depth + 1)
        right = generate_expression(depth + 1)
        return f"({left}{op}{right})"

# 数式の計算結果を得る関数
def evaluate_expression(expr):
    try:
        # evalを使って数式を評価し、四捨五入する
        result = eval(expr)
        return round(result, 2)  # 小数点以下2桁に丸める
    except ZeroDivisionError:
        return 'NaN'

# 数式を生成して計算結果を加える
def create_equation():
    expr = generate_expression()
    result = evaluate_expression(expr)
    return f"{expr}={result}"

# 100個の数式を生成してテキストファイルに保存
def generate_equations_to_file(num_equations, filename=None):
    if filename==None:
        for _ in range(num_equations):
            print(create_equation())
        return
    with open(filename, 'w') as file:
        for _ in range(num_equations):
            equation = create_equation()
            file.write(equation + '\n')
```

# eq03.txt

ポーランド記法から逆ポーランド記法に変換

1桁の自然数を扱うので、途中の空白は入れない

```python
import random
import operator

# 数式生成に使うオペレータと対応するPythonの演算子
operators = {
    '+': operator.add,
    '-': operator.sub,
    # '*': operator.mul,
    # '/': operator.truediv,
}

# ランダムな数式を生成する関数（中置記法）
def generate_expression(depth=0):
    if depth >= 1 and random.random() < 0.8:
        # 数字のみを返す
        return str(random.randint(0, 9))
    else:
        # 深さを増やしながら再帰的に数式を生成
        op = random.choice(list(operators.keys()))
        left = generate_expression(depth + 1)
        right = generate_expression(depth + 1)
        return f"({left}{op}{right})"

# 中置記法から前置記法（ポーランド記法）に変換する関数
def infix_to_prefix(expression):
    def precedence(op):
        if op == '+' or op == '-':
            return 1
        if op == '*' or op == '/':
            return 2
        return 0

    def apply_operator(operators, values):
        operator = operators.pop()
        right = values.pop()
        left = values.pop()
        values.append(f"{operator} {left} {right}")

    operators = []
    values = []
    i = 0
    while i < len(expression):
        if expression[i] == '(':
            operators.append(expression[i])
        elif expression[i].isdigit():
            val = ""
            while i < len(expression) and expression[i].isdigit():
                val += expression[i]
                i += 1
            values.append(val)
            continue
        elif expression[i] == ')':
            while len(operators) != 0 and operators[-1] != '(':
                apply_operator(operators, values)
            operators.pop()  # '(' を削除
        else:
            # 演算子の場合
            while (len(operators) != 0 and
                   precedence(operators[-1]) >= precedence(expression[i])):
                apply_operator(operators, values)
            operators.append(expression[i])
        i += 1

    # 残りの演算子をすべて適用
    while len(operators) != 0:
        apply_operator(operators, values)

    return values[-1]

# 構文木のノードを表現するクラス
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# 数式をポーランド記法として扱い、構文木を構築
def parse_polish(expression):
    tokens = expression.split()
    stack = []
    
    for token in reversed(tokens):
        node = Node(token)
        if token in operators:
            node.left = stack.pop()
            node.right = stack.pop()
        stack.append(node)
    
    return stack[0]  # 最終的な構文木

# 構文木から逆ポーランド記法に変換
def to_reverse_polish(node):
    if node is None:
        return []
    return to_reverse_polish(node.left) + to_reverse_polish(node.right) + [node.value]

# データセット用に式を生成
def generate_polish_and_reverse_polish():
    expr = generate_expression()  # ランダムな数式生成
    polish_expr = infix_to_prefix(expr)  # 中置記法から前置記法へ変換
    
    # ポーランド記法を逆ポーランド記法に変換
    tree = parse_polish(polish_expr)
    reverse_polish_expr = ' '.join(to_reverse_polish(tree))

    # polish_expr = polish_expr.replace(' ', '')  # スペースを削除
    # reverse_polish_expr = reverse_polish_expr.replace(' ', '')  # スペースを削除
    
    return f"{polish_expr} => {reverse_polish_expr}"

# ファイルに保存する関数
def save_equations_to_file(num_equations, filename=None):
    if filename == None:
        for _ in range(num_equations):
            print(generate_polish_and_reverse_polish())
        return
    with open(filename, 'w') as file:
        for _ in range(num_equations):
            equation = generate_polish_and_reverse_polish()
            file.write(equation + '\n')

# 例として10個の数式を生成
save_equations_to_file(10)
```

```python
save_equations_to_file(100000, 'data/eq03.txt')
```

# eq04-1.txt

- 10桁のランダムな自然数の偶奇を判定するタスク

```python
import random

# 10桁の自然数を生成
def generate_number():
    return str(random.randint(10**9, 10**10 - 1))

# 偶数か奇数かを判定
def is_even(number):
    return "0" if int(number[-1]) % 2 == 0 else "1"

# データセット用に10桁の自然数と偶奇を生成
def generate_number_and_label():
    number = generate_number()  # 10桁の自然数を生成
    label = is_even(number)  # 偶数か奇数かを判定
    return f"{number} => {label}"

# ファイルに保存する関数
def save_numbers_to_file(num_numbers, filename=None):
    if filename is None:
        for _ in range(num_numbers):
            print(generate_number_and_label())
        return
    with open(filename, 'w') as file:
        for _ in range(num_numbers):
            data = generate_number_and_label()
            file.write(data + '\n')

# 例として10個の自然数を生成して判定
save_numbers_to_file(10)
```
